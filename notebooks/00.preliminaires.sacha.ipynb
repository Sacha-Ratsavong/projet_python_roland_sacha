```python
import pandas as pd

df = pd.read_csv("data/raw/paris-2024-results-medals-oly-eng.csv", sep=";")

df.head()
```

Je cherche maintenant à créer un autre tableau qui compte le nombre de médailles par pays

```python
# On garde uniquement les lignes où il y a une médaille (Medal_type non vide)
# On ne garde que les colonnes utiles : Country, Country_code, Medal_type
df_medals = df[df["Medal_code"].notna()][["Country", "Country_code", "Medal_type"]]

# On regroupe par pays et par type de médaille, puis on compte le nombre de lignes
# .size() = compte le nombre de médailles par groupe
# .unstack() = transforme les types de médailles en colonnes (Gold / Silver / Bronze)
medals_by_country = (
    df_medals
    .groupby(["Country", "Medal_type"])
    .size()
    .unstack(fill_value=0)        # met 0 au lieu de NaN quand un pays n'a pas une médaille donnée
)

# On ajoute une colonne Total = somme des colonnes médailles
medals_by_country["Total"] = medals_by_country.sum(axis=1)

# On trie les pays du plus médaillé au moins médaillé
medals_by_country = medals_by_country.sort_values("Total", ascending=False)

medals_by_country.head()
```
